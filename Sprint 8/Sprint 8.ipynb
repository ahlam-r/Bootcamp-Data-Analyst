EJERCICIO 1 Correlación de todas las variables numéricas.

Heatmap de correlación muestra la matriz de correlación entre variables numéricas (amount, lat, longitude, weight, price).
Los valores cercanos a 1 indican alta correlación positiva, cercanos a -1 indican correlación negativa, y cerca de 0, ausencia de correlación. Weight y price tienen la mayor correlación positiva (0.19), aunque sigue siendo débil. Las demás correlaciones son prácticamente nulas, lo que indica baja relación lineal entre variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Seleccion columnas específicas numericas( no lo he hecho automaticamente con "(numeric_only=True)" porque me seleccionaba columnas de id)
cols = ['amount', 'lat', 'longitude', 'weight', 'price']
df_selected = df_full_merge[cols]

# Calcular matriz de correlación
corr_matrix = df_selected.corr()

# Crear el heatmap
sns.heatmap(data=corr_matrix, 
            annot=True, fmt='.2f', annot_kws={'size': 10},
            linewidths=1, cmap='coolwarm', vmax=1, vmin=-1)

plt.title("Correlación entre variables seleccionadas")
plt.tight_layout()
plt.show()
```

EJERCICIO 2 Implementa un jointplot.

Gráfico Jointplot entre amount y price. Combina un scatterplot con histogramas para observar la relación entre amount (eje x) y price (eje y).
Scatterplot: Los puntos están muy dispersos, lo que sugiere que no hay una relación lineal clara entre amount y price.
Histogramas: Amount tiene una distribución relativamente uniforme, con algunos picos. Price parece tener valores agrupados en ciertas franjas.

```python
import seaborn as sns
import matplotlib.pyplot as plt

#seleccionar solo las columnas necesarias
df_plot = df_full_merge[["amount", "price"]].dropna()

#Crear el jointplot
sns.jointplot(data=df_plot, x="amount", y="price", kind="scatter")

plt.suptitle("Relación entre Monto y Precio", y=1.02)
plt.show()
```
